```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error, f1_score, accuracy_score, roc_curve, roc_auc_score
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
X, y = load_boston(return_X_y=True)
X1 = np.hstack([np.ones(X.shape[0])[:, np.newaxis], X])
```

```python
X
```

```python
def print_regression_metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    print(f'MSE = {mse:.2f}, RMSE = {rmse:.2f}')

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg_f = LinearRegression(fit_intercept=False)
```

```python
lin_reg.fit(X, y)
print(lin_reg.coef_, lin_reg.intercept_, lin_reg.rank_)
y_pred = lin_reg.predict(X)
print_regression_metrics(y, y_pred)
```

```python
lin_reg.fit(X1, y)
print(lin_reg.coef_, lin_reg.intercept_, lin_reg.rank_)
y_pred = lin_reg.predict(X1)
print_regression_metrics(y, y_pred)
```

```python
lin_reg_f.fit(X, y)
print(lin_reg_f.coef_, lin_reg_f.intercept_, lin_reg_f.rank_)
y_pred_f = lin_reg_f.predict(X)
print_regression_metrics(y, y_pred_f)
```

```python
lin_reg_f.fit(X1, y)
print(lin_reg_f.coef_, lin_reg_f.intercept_, lin_reg_f.rank_)
y_pred_f = lin_reg_f.predict(X1)
print_regression_metrics(y, y_pred_f)
```
