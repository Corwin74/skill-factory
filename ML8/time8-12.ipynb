```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import time, datetime
```

```python
data = pd.read_csv('gold_clean.csv')
data = data.rename(columns = {'ts': 'unixtime'})
data['ts'] = pd.to_datetime(data.unixtime)
data = data[(data.ts >= '2018-01-10') & (data.ts <= '2018-01-11')]

fig = go.Figure(data  = [go.Candlestick(
            x=data['ts'],
            open  = data['open'],
            high  = data['high'],
            low   = data['low'],
            close = data['close']
        )])

fig.show()

data = data.reset_index(drop=True)
data.head()
```

```python
data['mr'] = data.close.rolling(window=100).mean()
```

```python
data[data.mr < data.close]
```

```python
data['std'] = data.close.rolling(window=100).std()
```

```python
data.iloc[99]
```

```python
data['upper'] = data.apply(lambda x: x.mr + x['std']*3, axis=1)
data['lower'] = data.apply(lambda x: x.mr - x['std']*3, axis=1)
```

```python
data.sample(10)
```

```python
data[data.close < data.lower].count()
```

```python
def myRollFunc(value):
    return (value.max()+value.min())/2
```

```python
data.close.rolling(window=50).apply(myRollFunc)
```

```python
data['r500'] = data.close.rolling(window=500).mean()
data['r100'] = data.close.rolling(window=100).mean()
```

```python
data['maza'] = data.apply(lambda x: np.sign(x.r500 - x.r100), axis=1)
```
